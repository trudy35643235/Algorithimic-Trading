Usually, any Sharpe ratio greater than 1.0 is considered acceptable to good by investors. A ratio higher than 2.0 is rated as very good. A ratio of 3.0 or higher is considered excellent. A ratio under 1.0 is considered sub-optimal

```python
import yfinance as yf
import pandas as pd

def hist_data(ticker_symbol, start_date):
    data = yf.download(ticker_symbol, start=start_date)
    data['Daily Return'] = data['Adj Close'].pct_change()
    return data.dropna()

def std_dev(data):
    return data.std()

def sharpe_ratio(data, risk_free_rate=0.0):
    
    # Calculate Mean 
    
    mean_daily_return = data['Daily Return'].mean()
    
    # Standard Deviation
    s = std_dev(data['Daily Return'])
    daily_sharpe_ratio = (mean_daily_return - risk_free_rate) / s
    sharpe_ratio = (252 ** 0.5) * daily_sharpe_ratio
    return sharpe_ratio

# Example usage
ticker_symbol = "SPRC"
start_date = "2023-11-01"
data = hist_data(ticker_symbol, start_date)
sharpe_ratio_value = sharpe_ratio(data)
print("Sharpe Ratio:", sharpe_ratio_value)
```
